```python
import requests

isbn = '9791190090018'
url = 'https://www.yes24.com/Product/Search?domain=ALL&query={}'
r = requests.get(url.format(isbn))
```

```python
print(r.text)
```

```python
from bs4 import BeautifulSoup
# 파서>> 입력 데이터를 받아 데이터 구조를 만드는 소프트웨어 라이브러리
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
prd_link = soup.find('a', attrs={'class':'gd_name'})
prd_link
```

```python
prd_link['href']
```

```python
url2 = 'https://www.yes24.com/'+prd_link['href']
r2 = requests.get(url2)
r2.text
```

```python
soup2 = BeautifulSoup(r2.text, 'html.parser')
prd_detail = soup2.find('div', attrs={'id':'infoset_specific'})
print(prd_detail)
```

```python
prd_tr_list = prd_detail.find_all('tr')
prd_tr_list
```

```python
for tr in prd_tr_list:
    if tr.find('th').get_text() == '쪽수, 무게, 크기':
        page_td = tr.find('td').get_text()
        break

print(page_td)
print(type(page_td))
```
