````python
import pandas as pd

ns_df = pd.read_csv('../data/namsan_04_2021.csv',dtype={'ISBN': str, '세트 ISBN': str, '주제분류번호': str})
ns_df.head()
````

````python
# loc << 인덱스 열위치, iloc << 인덱스의 위치
ns_book = ns_df.loc[:,'번호':'등록일자']
ns_book.head()
````

````python
selected_columns = ns_df.columns != 'Unnamed: 13'
selected_columns
````

````python
type(selected_columns)
````

````python
ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
````

````python
ns_book = ns_df.drop(['부가기호', 'Unnamed: 13'], axis=1)
ns_book.head()
````

````python
ns_book.drop('주제분류번호', axis=1, inplace=True)
ns_book.head()
````

````python
# DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
# how='any' NaN이 하나라도 있으면 삭제, how='all' 모든 값이 NaN이라면 삭제
ns_dropna = ns_df.dropna(axis=1, how='any')
ns_dropna.head()
````

````python
# DataFrame.duplicated(subset=None, keep='first')
# duplicated 메서드는 중복되는 행을 확인하는 메서드입니다.
# 행의 모든 요소가 동일한 행이 이미 존재할경우 해당 행은 True로 반환됩니다.
# https://wikidocs.net/154787

print(sum(ns_book.duplicated()))
print(ns_book.duplicated())
````

````python
print(ns_book.columns)
print(len(ns_book.columns))
````

````python
# 행의 모든 요소가 동일한 행이 이미 존재할경우 해당 행은 True로 반환됩니다.
sum(ns_book.duplicated(subset=['도서명', '저자', 'ISBN']))
````

````python
# subset : 특정 열만을 대상으로 할 수 있습니다. list의 사용도 가능합니다.
# keep : {first : 위부터 검사 / last : 아래부터 검사, False : 중복을 고려 안하고 무조건 중복이면 False} 검사 순서를 정합니다. 

dup_rows = ns_book.duplicated(subset=['도서명', '저자', 'ISBN'], keep=False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
````
